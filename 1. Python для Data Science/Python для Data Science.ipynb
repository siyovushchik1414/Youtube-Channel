# Что такое виртуальная среда в ЯП Python?

Виртуальная среда в Python - это инструмент, который помогает сохранять зависимости, требуемые различными проектами, отдельно, создавая для них изолированные виртуальные среды³. В корне своем, главная задача виртуальной среды Python – создание изолированной среды для проектов Python¹⁴. Это значит, что каждый проект может иметь свои зависимости, и зависимости одного проекта не влияют на зависимости другого⁴⁵. 

Виртуальное окружение (virtual environment) — это каталог, в который устанавливаются некоторые исполняемые файлы и скрипты. Среди файлов есть python для выполнения скриптов и pip для установки других модулей в окружении⁶.

Виртуальная среда Python позволяет управлять проектами изолированно друг от друга⁵. Это очень полезно, когда у вас есть два проекта: проект А и проект Б, которые оба имеют зависимость от одной и той же библиотеки – проект В. Проблема становится явной, когда мы начинаем запрашивать разные версии проекта В¹.

# Что такое pip в Python?

`pip` — это система управления пакетами, которая используется для установки и управления программными пакетами, написанными на Python⁵. Множество пакетов можно найти в Python Package Index (PyPI)⁵. 

Одно из главных преимуществ `pip` — это простота интерфейса командной строки, которая позволяет установить пакеты Python простой командой `pip install some-package-name`. Так же просто и удалять пакеты: `pip uninstall some-package-name`⁵. 

Важно, что `pip` предоставляет возможность управлять всеми пакетами и их версиями с помощью файла `requirements.txt`. Это позволяет эффективно воспроизводить весь необходимый список пакетов в отдельном окружении (например, на другом компьютере) или в виртуальном окружении. Это достигается с помощью правильно составленного файла `requirements.txt` и следующей команды: `pip install -r requirements.txt`⁵.

# Что такое компилятор?
Компилятор — это программа, которая переводит текст, написанный на языке программирования, в машинные коды¹². С помощью компиляторов компьютеры могут понимать разные языки программирования, в том числе высокоуровневые, то есть близкие к человеку и далекие от «железа»². Процесс работы компилятора с кодом называется компиляцией, или сборкой². 

По сути, компилятор — комплексный «переводчик», который собирает, или компилирует, программу в исполняемый файл². Исполняемый файл — это набор инструкций для компьютера, который тот понимает и может выполнить². Языки программирования, для перевода которых используются компиляторы, называются компилируемыми².

Без компилятора любой код на компилируемом языке программирования будет для компьютера просто текстом — он не распознает команды и не сможет их выполнить². Поэтому компилятор нужен, чтобы программы могли выполняться. Без него ничего не будет работать².

Так что можно сказать, что компилятор "переводит" то, что мы написали, на язык, который понимает компьютер.

# В чём разница между модулем и библиотекой?

**Модуль** в Python - это простой файл Python, который содержит коллекции функций и глобальных переменных и имеет расширение .py¹. Это исполняемый файл, и для организации всех модулей в Python есть понятие пакета¹.

**Библиотека** в Python - это набор связанных функций кода, которые позволяют выполнять множество задач без написания своего кода¹. Это повторно используемый фрагмент кода, который мы можем использовать, импортировав его в нашу программу¹. Мы можем просто использовать его, импортировав эту библиотеку и вызвав метод этой библиотеки с помощью точки¹.

Таким образом, модуль - это отдельный файл с определениями функций и переменных, а библиотека - это набор таких модулей, которые вместе предоставляют определенные функциональные возможности¹².

# Функция

# Что такое функция?
Функция в программировании - это блок кода, который выполняет определенную задачу и может принимать параметры и возвращать результат.

Функция в программировании - это как рецепт приготовления блюда. В этом рецепте есть определенные шаги (код внутри функции), которые нужно выполнить, чтобы получить конечный результат (возвращаемое значение функции). 

Также, как и в рецепте, где нам нужны ингредиенты (аргументы или параметры функции), чтобы приготовить блюдо. Эти "ингредиенты" мы передаем функции, когда вызываем ее.

Вот очень простой пример функции на языке Python:

```python
def greet(name):
    return "Привет, " + name + "!"
```

Здесь `greet` - это имя функции, `name` - это "ингредиент" или параметр, который мы передаем функции, а `"Привет, " + name + "!"` - это результат, который функция возвращает после выполнения²⁴.

````python
# n! = 1 * 2 * 3 * ... * n

def factorial(n):
    if(n == 0):
        return 1
    while(n > 0):
        return n * factorial(n - 1)

print(factorial(10))

# 1 * 2 * 3 = 2 * 3 = 6
````

# Исключения

Исключения в Python - это еще один тип данных, который необходим для сообщения программисту об ошибках¹². Вот несколько примеров исключений:

1. Деление на ноль:

```python
>>> 100 / 0

```

```python
>>> 2 + '1'
```

# Почему нельзя делить на ноль и что такое неопределённость в математике?

0 / 0 = 1, inf, 0
X / X = 1
X / 0 = inf
0 / X = 0

Деление на ноль в математике является недопустимым, поскольку не существует числа, которое при умножении на ноль даст что-то, отличное от нуля. Если бы мы могли делить на ноль, то результатом такого деления могла бы быть бесконечность или неопределенность.

Неопределенность в математике относится к ситуациям, когда результат операции не ясен или не может быть точно определен. Например, выражения типа 0/0 или ∞/∞ являются неопределенностями. Это потому, что нет конкретного числа, которое можно было бы однозначно присвоить этим выражениям. В математике существуют методы для "раскрытия" или обработки этих неопределенностей, такие как правило Лопиталя.


# Списки
Списки в Python - это упорядоченные изменяемые коллекции объектов произвольных типов¹². В других языках программирования аналогичная структура данных называется массивами⁴. Списки могут содержать любое количество любых объектов, включая другие списки¹. 

Создать список можно несколькими способами¹:
- Обработать любой итерируемый объект (например, строку) встроенной функцией `list`¹.
- Использовать литералы для создания списка¹.
- Использовать генераторы списков¹.

Списки в Python поддерживают множество методов, таких как `append`, `extend`, `insert`, `remove`, `pop`, `index`, `count`, `sort`, `reverse`, `copy`, `clear`¹. Эти методы позволяют выполнять различные операции над списками, такие как добавление элементов, удаление элементов, сортировка и другие¹.

Важно отметить, что методы списков изменяют сам список¹. Это отличает их от строковых методов, которые возвращают новую строку и не изменяют исходную¹.

# Что такое литералы?

# Что такое словарь defaultdict?
`Defaultdict` в Python - это подкласс класса `dict`, который предоставляет значение по умолчанию для ключа, которого нет в словаре¹. Это означает, что если вы попытаетесь получить доступ или изменить отсутствующий ключ, `defaultdict` автоматически создаст ключ и сгенерирует для него значение по умолчанию³. Это делает `defaultdict` ценным вариантом для обработки отсутствующих ключей в словарях³.

Вот пример использования `defaultdict` в Python:
```python
from collections import defaultdict

# Функция для возврата значения по умолчанию
def default_value():
    return "This key is not present"

# Создание defaultdict
dict_1 = defaultdict(default_value)
dict_1["ABC"] = 1
dict_1["DEF"] = 2

print(dict_1["ABC"])  # Выведет: 1
print(dict_1["XYZ"])  # Выведет: This key is not present
```
В этом примере, если мы попытаемся получить доступ к ключу, которого нет в `dict_1` (например, "XYZ"), `defaultdict` автоматически создаст этот ключ с значением, возвращаемым функцией `default_value`¹.

># <center>Инструкция assert<center>

Инструкция `assert` в Python используется для проверки истинности условия². Если условие истинно, то выполнение программы продолжается. Если условие ложно, то генерируется исключение `AssertionError`¹².

В Python есть две формы инструкции `assert`¹:

1. **Базовая форма**:
```python
assert expression
```
Это эквивалентно следующему коду¹:
```python
if __debug__:
    if not expression:
        raise AssertionError
```
2. **Расширенная форма**:
```python
assert expression1, expression2
```
Это эквивалентно следующему коду¹:
```python
if __debug__:
    if not expression1:
        raise AssertionError(expression2)
```
В расширенной форме `expression2` используется в качестве аргумента для `AssertionError`, если `expression1` оказывается ложным¹.

Вот пример использования инструкции `assert`:

```python
def divide(x, y):
    assert y != 0, "Деление на ноль недопустимо"
    return x / y

print(divide(10, 2))  # Вывод: 5.0
print(divide(10, 0))  # Вызывает AssertionError: Деление на ноль недопустимо
```
В этом примере функция `divide` использует инструкцию `assert` для проверки, что делитель `y` не равен нулю. Если `y` равен нулю, то генерируется исключение `AssertionError` с сообщением "Деление на ноль недопустимо".

# Что за __debug__?
`__debug__` в Python - это встроенная константа. Она имеет значение `True`, если Python был запущен без опции `-O`. Значение `__debug__` не может быть изменено во время выполнения Python.

Одно из основных применений `__debug__` - это использование вместе с инструкцией `assert`. Инструкция `assert` выполняется только тогда, когда `__debug__` имеет значение `True`. Если Python запущен с опцией `-O`, то `__debug__` будет `False`, и инструкции `assert` будут игнорироваться.

Опция `-O` в Python используется для оптимизации кода. Когда Python запускается с этой опцией, он выполняет некоторые дополнительные оптимизации, такие как удаление инструкций `assert` и установка встроенной константы `__debug__` в `False`. Это может быть полезно для увеличения скорости выполнения программы и уменьшения ее размера. Однако следует быть осторожным при использовании этой опции, поскольку она может изменить поведение программы.

````python
# assert 1 + 1 == 3

# assert 1 + 1 == 3, "Это выражение не верно"

def square(x):
    return x ** 2

assert square(5) == 25, "Квадрат неверен"
````

# Что такое ООП и почему оно так называется?
Объектно-ориентированное программирование (ООП) - это подход к программированию, где программы структурируются в виде объектов - экземпляров классов, которые могут содержать данные и код: данные в виде полей (часто известных как атрибуты), а код - в виде процедур (часто известных как методы)¹. ООП называется так, потому что оно ориентировано на объекты, а не на логику¹. Этот подход поддерживается многими языками программирования, включая Python, Java, C++, Ruby, JavaScript и многими другими¹.

В Python классы определяются с помощью ключевого слова `class`¹. Вот пример простого класса в Python¹:

```python
class MyClass:
    x = 5
```
В этом примере `MyClass` имеет атрибут `x`, который равен 5¹. Вы можете использовать класс для создания объектов¹:

```python
p1 = MyClass()
print(p1.x)
```
В этом примере `p1` является объектом (или экземпляром) класса `MyClass`, и мы выводим значение его атрибута `x`¹.

В Python `self` и `__init__` являются ключевыми словами, используемыми в контексте классов и объектов¹²³.

- `self` представляет экземпляр класса¹²³. Это ссылка на текущий объект, и с его помощью можно получить доступ к атрибутам и методам класса¹²³. В большинстве объектно-ориентированных языков это передается как скрытый параметр методам, определенным на объекте². В Python это необходимо объявлять явно².

- `__init__` - это специальный метод, который вызывается при создании объекта¹²³. Он известен как конструктор в объектно-ориентированных концепциях¹²³. Этот метод используется для инициализации атрибутов класса¹²³.

Вот пример использования `self` и `__init__` в Python²:

```python
class MyClass:
    def __init__(self):
        self.my_attribute = "Hello"

    def my_method(self):
        print(self.my_attribute)

obj = MyClass()
obj.my_method()  # Выводит: Hello
```
В этом примере `MyClass` имеет метод `__init__`, который устанавливает атрибут `my_attribute` для каждого созданного объекта². Затем метод `my_method` использует `self` для доступа к этому атрибуту²..

# 3 основных принципа ООП: Инкапсуляция, полиморфизм, наследование. Что это такое?
Инкапсуляция, наследование и полиморфизм являются основными принципами объектно-ориентированного программирования (ООП). Давайте рассмотрим каждый из них на примере Python.

Инкапсуляция:
Инкапсуляция - это механизм, который ограничивает доступ к данным и методам внутри объекта. В Python инкапсуляция обычно реализуется с помощью префиксов "__" перед именами атрибутов или методов, что делает их "приватными". Например:
```python
class Car:
    def __init__(self, brand, model):
        self.__brand = brand
        self.__model = model

    def get_brand(self):
        return self.__brand

    def get_model(self):
        return self.__model

car = Car("Toyota", "Corolla")
print(car.get_brand())  # Выведет "Toyota"
print(car.__brand)  # Ошибка, так как атрибут приватный
```

Наследование:
Наследование позволяет создавать новый класс на основе существующего. В Python класс наследуется путем указания базового класса в скобках после имени нового класса. Например:
```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def get_brand(self):
        return self.brand

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

car = Car("Toyota", "Corolla")
print(car.get_brand())  # Выведет "Toyota"
```

Здесь создается новый класс Car, который наследуется от базового класса Vehicle. В методе __init__ нового класса вызывается метод __init__ базового класса с помощью функции super(). Это позволяет использовать функциональность базового класса в новом классе, а затем добавляется новый атрибут model. Таким образом, при создании объекта класса Car, он будет иметь как атрибут brand из базового класса Vehicle, так и атрибут model из класса Car.

Полиморфизм:
Полиморфизм позволяет использовать один и тот же метод для разных типов данных. В Python полиморфизм достигается за счет динамической типизации. Например:
```python
class Cat:
    def make_sound(self):
        print("Meow")

class Dog:
    def make_sound(self):
        print("Woof")

def animal_sound(animal):
    animal.make_sound()

cat = Cat()
dog = Dog()

animal_sound(cat)  # Выведет "Meow"
animal_sound(dog)  # Выведет "Woof"
```

Это основные принципы ООП в Python. Используя эти принципы, можно создавать более структурированный и гибкий код.
