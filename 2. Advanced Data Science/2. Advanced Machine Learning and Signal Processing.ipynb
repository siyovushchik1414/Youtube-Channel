Pipelines are a concept that allows for the combination of data from various sources into a single dataset. They can be represented as a list of necessary transformation steps, which can then be sequentially applied to a dataset. The machine learning framework, scikit-learn, heavily utilizes pipelines. However, scikit-learn has a significant limitation: if the memory on your computer is exceeded, the program may not run.

This is where Apache Spark comes into play. Apache Spark is an open-source library for scientific computing. It's a parallel data framework that distributes data across a system of cluster nodes. The more cluster nodes you have, the better it scales. For instance, doubling the number of nodes will nearly double the performance. Apache Spark employs the Spark ML framework.

In addition to scikit-learn, we will also discuss System ML, a machine learning library. System ML is an optimizer for linear algebra and serves as a SQL (Structured Query Language) engine for linear algebra.

# Linear Algebra
Let us start by considering the necessary concepts.

**Scalar**

A scalar is a number that can be used to represent a value.
Examples of scalars are: 1, 2, 3, 4.2, 0.5, 6, 7 and so on.  

**Vector** 

A vector is a set of scalars that can be used to represent a vector. (1,7,2)
Tuple. A tuple is a vector that can have elements of different types. (1,7,2,3.6)

**Matrix** 

A matrix is a set of vectors of the same length. For example, if we group vector a(1,7,2) and vector b(3.6,4.2,0.5) to a matrix, then the matrix can be represented as (1, 7, 2)(3.6, 4.2, 0.5). A matrix has n columns and m rows. If n is equal to m, then the matrix is called a square matrix.

**Tensor** 

A tensor is a N-dimensional matrix. Let us try to imagine a 3-dimensional tensor. Basically, a 3-dimensianal tensor is cube. A cube has n rows, n columns, and n dimensions. That is an easy way to imagine a 3-dimensional tensor. But it is not easy to imagine a 4-dimensional tensor. 



**Vector Multiplication**

A vector can be multiplied by a scalar. For example, the vector (1,7,2) can be multiplied by a scalar 3.6.
Vector also has a method called dot() that can be used to calculate the dot product of two vectors.
Dot product of two vectors is the sum of the dot products of their elements.
Generally speaking, a vector multiplication is a linear combination of those.



** Vector Matrix Multiplication**

A vector can be multiplied by a matrix. For example, the vector (1,7,2) can be multiplied by the matrix (3.6,4.2,0.5).
The result of the multiplication is a new vector.
